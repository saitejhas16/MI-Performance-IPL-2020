### STATS AND PERFORMANCE OF MUMBAI INDIANS IN IPL 2020

```python
#importing pandas,matplotlib and numpy
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#loading data 
path=r"C:\Users\SAI\OneDrive\Desktop\Ipl_2020.xlsx"
df=pd.read_excel(path)
#viewing top 5 rows of the data
df.head(5)
```

```python
##viewing the columns
df.columns
```

```python
##removing unwanted columns and creating a data set
cols=[2,4,6,8,9,12,14,15,16,18]
df_mod=df.drop(columns=[df.columns[i] for i in cols])
df_mod.head()
```

```python
##filtering only MI matches from the data
df_mi=df_mod[df_mod['Match'].str.contains('MI')]
df_mi
```

```python
#changing values of result column based on MI's result
df_mi.loc[:, 'Result'] = np.where(df_mi['winner'] == 'MI', 'Won', 'Lost')
df_mi
```

```python
#lets remove date and time column
df_mi=df_mi.drop(columns=['Date','Time India (IST)'])
df_mi
```

```python
#saving the cleaned data set
df_mi.to_excel("Final.xlsx")
```

## PIE CHART FOR NUMBER OF MATCHES WON VS LOST BY MI

```python
# Calculate count of results
result_counts = df_mi['Result'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(6,4))
result_counts.plot(kind='pie', autopct='%1.1f%%', colors=['red', 'green'])
plt.title('Proportion of Matches Won and Lost')
plt.show()
```

## Bar Plot by Venue: Count of Wins and Losses by Venue

```python
# Count of wins and losses by venue
venue_result_counts = df_mi.groupby(['Venue', 'Result']).size().unstack().fillna(0)

# Plotting
venue_result_counts.plot(kind='bar', stacked=True, figsize=(6,5))
plt.title('Count of Matches Won and Lost by Venue')
plt.xlabel('Venue')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Result')
plt.show()
```
